```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
x = np.random.laplace(size=n)

width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x))

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor='black')
plt.ylabel('N per bins')
plt.xlabel('x')

plt.savefig('Prompt-10-Response.pdf',bbox_inches='tight',facecolor='white')
```
