```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1.,100)
```

```python
def exp(x):
    return np.exp(x)
```

```python
y = exp(x)
f = plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.xlabel('Time (ms)')
plt.ylabel('Awesomeness (rads)')
plt.savefig('Average ASTR-19 Lecture',bbox_inches='tight',facecolor='white')
```
