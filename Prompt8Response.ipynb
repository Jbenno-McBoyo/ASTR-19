```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1.,100)

def sin(x):
    return np.sin(x)
def cos(x):
    return np.cos(x)
```

```python
f, axarr = plt.subplots(1,2)

axarr[0].plot(x,sin(x))
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title('x v. sin(x)')
axarr[0].set_aspect(1)

axarr[1].plot(x,cos(x))
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title('x v. cos(x)')
axarr[1].set_aspect(2)

f.subplots_adjust(wspace=0.5)

plt.savefig('sin-and-cos.png',bbox_inches='tight',facecolor='white')
```
